```python
from datasets import Dataset
from PIL import Image
import os
import json

# JSON íŒŒì¼ ê²½ë¡œ
json_path = "hyundaicar_descript_merge_all.json"
image_root = ""  # ì´ë¯¸ì§€ ê²½ë¡œê°€ ìƒëŒ€ê²½ë¡œì¼ ê²½ìš° ê¸°ì¤€ ë””ë ‰í† ë¦¬

# JSON ë¡œë”©
with open(json_path, 'r') as f:
    data = json.load(f)

# ì´ë¯¸ì§€ ê²½ë¡œ + ì„¤ëª… ì¶”ì¶œ
samples = []
for item in data:
    image_path = os.path.join(image_root, item["image_path"])
    text = item["description"]
    samples.append({"image": image_path, "text": text})
    
# ðŸ¤– ðŸ¤ diffusersì—ì„œ ìš”êµ¬í•˜ëŠ” datasets.Dataset í˜•íƒœë¡œ ë³€í™˜
dataset = Dataset.from_list(samples)
dataset
```
